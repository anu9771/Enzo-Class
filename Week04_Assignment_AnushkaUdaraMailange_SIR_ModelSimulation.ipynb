```python
#Defining SIR Model
#Inputs are Number of suseptible people (S), Number of Infected people (I), Number of recovered people (R), alpha, gamma, and population size (N)
#It returns rate of changeing suseptible, infectious, and recovered people ds/dt, di/dt and dr/dt

def sir_model(S, I, R, alpha, gamma, N):

    dS = -alpha * S * I / N
    dI = alpha * S * I / N - gamma * I
    dR = gamma * I
    return dS, dI, dR

#Defining Euler method
#Inputs are Number of suseptible people (S), Number of Infected people (I), Number of recovered people (R), alpha, gamma, and population size (N)
#Also time step and total number of steps(days)
#For each successive day, calculated S,I,R values are appended to a list

def euler_method(S, I, R, alpha, gamma, N, dt, steps):

    results = []
    for _ in range(steps):
        dS, dI, dR = sir_model(S, I, R, alpha, gamma, N)
        S += dS * dt
        I += dI * dt
        R += dR * dt
        results.append((S, I, R))
    return results

# Parameters
N = 1000
S0 = 999
I0 = 1
R0 = 0
alpha = 0.3
gamma = 0.1
dt = 1.0
steps = 160

# Run the simulation
time_series = euler_method(S0, I0, R0, alpha, gamma, N, dt, steps)

# Print the results
#emurate was used to set a counter
for day, (S, I, R) in enumerate(time_series):
    print(f"Day {day+1}: S={S:.2f}, I={I:.2f}, R={R:.2f}")
```

```python
import matplotlib.pyplot as plt


days = range(steps)
S_values = [S for S, I, R in time_series]
I_values = [I for S, I, R in time_series]
R_values = [R for S, I, R in time_series]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(days, S_values, label='Susceptible (S)', color='blue')
plt.plot(days, I_values, label='Infected (I)', color='red')
plt.plot(days, R_values, label='Recovered (R)', color='green')

# Add labels and title
plt.xlabel('Days')
plt.ylabel('Number of Individuals')
plt.title('SIR Model Simulation')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
